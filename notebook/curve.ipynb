 #### BSA-PM: Position Mutation
 #### Non-linear parameter decrease from 2 to 0.
 #### a = a_std - (a_std - a_fin) * (1/(e-1) * (e^(FEs/Max_FEs) - 1))^2
 #### a_std = 2, a_fin = 0

```python

# Function definition
def curve_function(FEs, Max_FEs):
    return 2 - 2 * (1 / (np.e - 1) * (np.exp(FEs / Max_FEs) - 1)) ** 2

# Parameters
Max_FEs = 500
FEs_values = np.linspace(0, Max_FEs, 1000)

# Generate curve values
curve_values = curve_function(FEs_values, Max_FEs)

# Plotting the curve
plt.plot(FEs_values, curve_values, label='Corrected Curve Function')
plt.title('Corrected Curve Function')
plt.xlabel('FEs (Function Evaluations)')
plt.ylabel('a')
plt.legend()
plt.grid(True)
plt.show()
```

#### Mutation Step
#### p = C_max - rand * (-(sin(pi/(2*FEs)) + C_min))
#### C_max = 3, C_min = 1

```python
# Function definition for the new curve
def new_curve_function(FEs, C_max, C_min):
    rand = np.random.random(len(FEs))
    return C_max - rand * (-(np.sin(np.pi / (2 * FEs)) + C_min))

# Parameters
C_max = 3
C_min = 1

# Generate new curve values
new_curve_values = new_curve_function(FEs_values, C_max, C_min)

# Plotting the new curve
plt.plot(FEs_values, new_curve_values, label='New Curve Function')
plt.title('New Curve Function')
plt.xlabel('FEs (Function Evaluations)')
plt.ylabel('p')
plt.legend()
plt.grid(True)

# Marking dots for FEs = 0, 1, 2, 3, ..., 500
dots_FEs = np.arange(0, 501, 50)
dots_values = new_curve_function(dots_FEs, C_max, C_min)
plt.scatter(dots_FEs, dots_values, color='red', label='Dots at FEs')

plt.legend()
plt.show()
```

```python
# Corrected function definition for the new curve with rand in the range [0, 1)
def corrected_new_curve_function(FEs, C_max, C_min):
    rand = np.random.rand(len(FEs))  # Use np.random.rand for random numbers in [0, 1)
    return C_max - rand * (-np.sin(np.pi / (2 * FEs)) + C_min)

# Generate corrected new curve values
corrected_new_curve_values = corrected_new_curve_function(FEs_values, C_max, C_min)

# Plotting the corrected new curve
plt.plot(FEs_values, corrected_new_curve_values, label='Corrected New Curve Function')
plt.title('Corrected New Curve Function')
plt.xlabel('FEs (Function Evaluations)')
plt.ylabel('p')
plt.legend()
plt.grid(True)

# Marking dots for FEs = 0, 1, 2, 3, ..., 500
dots_FEs = np.arange(0, 501, 10)
dots_values = corrected_new_curve_function(dots_FEs, C_max, C_min)
plt.scatter(dots_FEs, dots_values, color='red', label='Dots at FEs')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the concave function and chord
def concave_function(x):
    return 1/2 * x ** 4 

def chord(x):
    return 1 - x

# Generate x values
x_values = np.linspace(0, 1, 100)

# Calculate corresponding y values for the function and chord
y_values_function = concave_function(x_values)
y_values_chord = chord(x_values)

# Plot the concave function and chord
plt.plot(x_values, y_values_function, label='Concave Function: $f(x) = -x^2 + 2x$')
plt.plot(x_values, y_values_chord, '--', label='Chord: $y = 1 - x$')
plt.scatter([0, 1], [1, 0], color='red', marker='o', label='Points (0,1) and (1,0)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Concave Function Passing Through (0,1) and (1,0)')
plt.grid(True)
plt.show()
```
